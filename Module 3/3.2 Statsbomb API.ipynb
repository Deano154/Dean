# Statsbomb API

Statsbomb has a public API where they have released event data from a lot of different leagues, matches, and seasons. They also have their Statsbomb360 Data which is a hybrid of tracking data and event data. It bascially allows you to get more information about event data and where other players are on the field at the time of the event.

For example, you can get data from the past couple of world cups, Euros, they even have all the data from every match Messi has played in.

We will be primarily using the Statsbomb API in the visualizations section & data analysis section of this course.

Statsbomb makes it a little bit difficult at first to find the data we need mainly because you need to know the competition_id and the season_id to get the data you want.

```python
from statsbombpy import sb
```

```python
# Load all competitions
sb.competitions()

# Take note of the competition_id and season_id
```

```python
# Lets get more information about the recent African Cup of Nations
sb.matches(competition_id=1267, season_id=107)

# Take note of the match_id
```

```python
# We can get one match like this:
sb.events(match_id=3923881).head(10)
```

```python
# If we wanted to get all the matches we need to loop through the match_ids
all_matches = []
for match_id in sb.matches(competition_id=1267, season_id=107).match_id:
    temp_match = sb.events(match_id=match_id)
    all_matches.append(temp_match)
```

```python
# Then we can concatenate all the matches into one dataframe
import pandas as pd
df = pd.concat(all_matches)
```

##### A couple of notes:
- You probably won't need to get all of the data from all of the matches unless you are doing larger analysis on a specific team or player.
- The statsbomb api can be somewhat slow fetching all of the data and if you are gathering a lot of data you might end up with a large dataframe
- You can use filtering to get just the data for a specific player you want to analyze for example:

```python
df.player.unique()
```

```python
df = df[df['player'] == 'Victor James Osimhen'] # Now we only have data for Victor Osimhen
```
